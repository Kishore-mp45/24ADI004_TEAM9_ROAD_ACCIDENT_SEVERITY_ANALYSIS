```python
import pandas as pd
import numpy as np
import random

# Your list of replacement cities
replacement_cities = ['Wasta', 'Okaton', 'Adell','Masardis', 'Saint Croix']

# Find indices where 'city' is null
null_indices = df_1[df_1['City'].isnull()].index

# Generate a random list of 253 choices from your list
random_choices = [random.choice(replacement_cities) for _ in range(len(null_indices))]

# Fill the nulls
df_1.loc[null_indices, 'City'] = random_choices
```

Handling the wind-direction


```python
df_1['Wind_Direction'] = df["Wind_Direction"]
df_1["Wind_Direction"].isnull().sum()
```

```python
valid_dirs = [
    'N','NNE','NE','ENE','E','ESE','SE','SSE',
    'S','SSW','SW','WSW','W','WNW','NW','NNW',
    'CALM','VAR'
]

df_1.loc[~df['Wind_Direction'].isin(valid_dirs), 'Wind_Direction'] = np.nan
```

```python
df_1['Wind_Direction'].fillna(df_1['Wind_Direction'].mode()[0], inplace=True)
```

```python
df_1['Weather_Condition']=df_1['Weather_Condition'].fillna('Fair')
```

```python
df_1['Humidity(%)'] = df_1.groupby('State')['Humidity(%)'] \
                      .transform(lambda x: x.fillna(x.median()))
```

Handling Visibilty

```python
df_1['Visibility(mi)'] = df_1.groupby('Weather_Condition')['Visibility(mi)'] \
                          .transform(lambda x: x.fillna(x.median()))
```

```python
df_1['Pressure(in)'] = df_1.groupby(['State'])['Pressure(in)'] \
                       .transform(lambda x: x.fillna(x.median()))
```

Z-Score

```python
import scipy
from scipy.stats import zscore


z_cols = [
    'Temperature(F)',
    'Wind_Chill(F)',
    'Pressure(in)',
    'Wind_Speed(mph)'
]

# compute z-scores
z_scores = np.abs(zscore(df_1[z_cols]))

# relaxed threshold (important for skewed data)
threshold = 5

# filter rows
df_1 = df_1[(z_scores < threshold).all(axis=1)]

print("After Z-score filtering:", df_1.shape)
```

IQR Outlier Filtering

```python
def iqr_filter(df, col):
    Q1 = df_1[col].quantile(0.25)
    Q3 = df_1[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return df_1[(df[col] >= lower) & (df_1[col] <= upper)]
```

```python
iqr_cols = [
    'Distance(mi)',
    'Visibility(mi)',
    'Precipitation(in)',
    'Humidity(%)'
]

for col in iqr_cols:
    df_1 = iqr_filter(df_1, col)

print("After IQR filtering:", df_1.shape)
```
