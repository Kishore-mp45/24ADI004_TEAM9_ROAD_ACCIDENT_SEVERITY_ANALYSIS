Z-Score

```python
import scipy
from scipy.stats import zscore


z_cols = [
    'Temperature(F)',
    'Wind_Chill(F)',
    'Pressure(in)',
    'Wind_Speed(mph)'
]

# compute z-scores
z_scores = np.abs(zscore(df_1[z_cols]))

# relaxed threshold (important for skewed data)
threshold = 5

# filter rows
df_1 = df_1[(z_scores < threshold).all(axis=1)]

print("After Z-score filtering:", df_1.shape)
```

IQR Outlier Filtering

```python
def iqr_filter(df, col):
    Q1 = df_1[col].quantile(0.25)
    Q3 = df_1[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return df_1[(df[col] >= lower) & (df_1[col] <= upper)]
```

```python
iqr_cols = [
    'Distance(mi)',
    'Visibility(mi)',
    'Precipitation(in)',
    'Humidity(%)'
]

for col in iqr_cols:
    df_1 = iqr_filter(df_1, col)

print("After IQR filtering:", df_1.shape)
```

Sanity Threshold Filtering

```python
df_1 = df_1[df_1['Wind_Speed(mph)'] < 200]
df_1 = df_1[df_1['Temperature(F)'] > -100]
df1 = df_1[df_1['Visibility(mi)'] < 200]
```
